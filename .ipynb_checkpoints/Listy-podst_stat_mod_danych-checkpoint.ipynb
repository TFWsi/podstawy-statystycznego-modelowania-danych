* **Air.Flow** - Przepływ powietrza,

* **Acid.Conc.** - Koncentracja kwasu,

### i) Określ model liniowy zależności między zmienną stack.loss a temperaturą wody.

#### Model 1

```r
model <- lm(stack.loss~Water.Temp+Acid.Conc.)
summary(model)
```

### Interpretacja

W podanym modelu jedynie zmienna **Water.Temp** wykazuje istotność statystyczną. Można więc odrzucić z niego zmienną **Acid.Conc.**

### Interpretacja

Zmienna **Water.Temp** istotnie wpływa na ogólną efektywność rośliny. Współczynnik dopasowania $R^2$ określa stopień wyjaśnienia danych przez model na poziomie 0.75.

$stack.loss = -41.91 + 2.82 \cdot Water.Temp$

```r
plot(Water.Temp, stack.loss, pch=19)
abline(model1, col='red')
grid()
legend("topleft", c("regresja dla wszystkich danych"), col = c("red"),
                    lty = c(1), pch = c(NA), lwd = c(1))
```

### ii) Czy któraś z obserwacji może być uznana za obserwację o dużej dźwigni?

### iii) Jeśli tak, przedstaw na wykresie prostą regresji dopasowaną do wszystkich obserwacji oraz prostą dopasowaną po usunięciu tego punktu.

```r
summary(model2)
```

### Interpretacja

Usunięcie punktu o dużej dźwigni zmniejszyło błąd standardowy reszt o około 0.3, jednak jednocześnie współczynnik dopasowania modelu $R^2$ spadł z poziomu 0.754 do poziomu 0.746.

### Przeprowadź analizę dyskryminacyjną korzystając z funkcji lda w pakiecie MASS. Prawdopodobieństwa a priori przynależności do klas przyjmij jako równe 0.5.

```r
model_lda = lda(owner ~ lotsize + income, data = kosiarki, prior = c(0.5, 0.5))
model_lda
```

### i) Jakie są średnie wartości w grupach?

```r
model_lda$means
```

### ii) Jakie są wartości współczynników funkcji dyskryminacyjnych?

```r
model_lda$svd
```

### iii) Dla każdej z funkcji dyskryminacyjnych określ która ze zmiennych objaśniających jest najbardziej istotna.

Posiadamy jedynie jedną funkcję dyskryminacyjną, dla której bardziej istotna jest zmienna objaśniająca **income** ze względu na niższą wartość współczynnika.

### iv) Jakie są średnie wartości funkcji dyskryminacyjnych w każdej z grup?

```r
LD1 = predict (model_lda)$x 
(LD1_1 = mean(LD1[kosiarki$owner == 1]))
(LD1_2 = mean(LD1[kosiarki$owner == 2]))
```

```r
m.lda.cv <- lda(type~., data=df, prior=c(0.5,0.5),CV=TRUE)
m.qda.cv <- qda(type~., data=df, prior=c(0.5,0.5),CV=TRUE)
```

### Dla danych DANE_wino.txt skonstruować drzewo klasyfikacyjne

```r
drzewo <- rpart(as.factor(V1) ~ ., data = wino, minsplit = 5, cp=0.0001)
plot(drzewo)
text(drzewo, use.n = TRUE)
```

### I. Wybrać optymalną wartość współczynnika złożoności i wyznaczyć ponownie drzewo z uzyskaną wartością współczynnika złożoności.

```r
plotcp(drzewo)
```

Optymalny współczynnik złożoności wynosi 0.000097.

```r
drzewo2 <- prune.rpart(drzewo, cp = 0.000097)
plot(drzewo2)
text(drzewo2, use.n = TRUE)
```

### II. Wybrać drzewo optymalne przy pomocy metody kros-walidacji.

```r
n = nrow(wino)
wyniki = rep(0, n)
wyniki2 = rep(0, n)

for (i in 1:n) {
    train = wino[-i, ]
    test = wino[i, ]
    drzewo_1 = rpart(factor(V1) ~ ., data = train, minsplit = 5, cp = 0.0001)
    drzewo_2 = rpart(factor(V1) ~ ., data = wino, minsplit = 5, cp = 0.000097)
    pred_1 = predict(drzewo_1, newdata = test, type = "class")
    pred_2 = predict(drzewo_2, newdata = test, type = "class")   
    if (pred_1 != factor(test$V1))
        wyniki[i] = 1
    if (pred_2 != factor(test$V1))
        wyniki2[i] = 1
}
```

```r
sum(wyniki) / n * 100
sum(wyniki2) / n * 100
```

W przypadku kroswalidacji bardziej optymalne jest drzewo 2.

### III. Następnie oszacować ułamek poprawnych klasyfikacji oraz błąd klasyfikacji na podstawie kros-walidacji.

```r
paste('Drzewo 1:',(sum(wyniki) / n * 100))
paste('Drzewo 2:',(sum(wyniki2) / n * 100))
```

Błąd klasyfikacji wynosi 11% w przypadku drzewa 1, oraz 1% w przypadku drzewa 2. Poprawnie zaklasyfikowane zostanie odpowiednio 89 oraz 99 procent obserwacji.

```r
drzewo3 <- rpart(as.factor(V1) ~ ., data = wino, minsplit = 5, cp=0.005)
plot(drzewo3)
text(drzewo3, use.n = TRUE)
```
